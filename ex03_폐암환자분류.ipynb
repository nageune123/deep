```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/data/ThoraricSurgery.csv',
                   header=None)
data.head()
```

```python
# 문제와 정답 분리
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
# 훈련과 평가로 분리
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size =0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# 중간층은 3개의 층으로 구성, activation: relu,units : 마음대로

from tensorflow.keras import Sequential # 딥러닝 모델의 뼈대
from tensorflow.keras.layers import InputLayer, Dense # 하나의 층의 디자인



model=Sequential()
# 입력층
#model.add(InputLayer(input_shape=(17,))) # 하나씩의 데이터를 입력 받음
#model.add(Dense(units= 48, activation='relu')) 

# 입력층
model.add(Dense(units= 48, activation='relu',input_dim = 17)) 
# 중간층
model.add(Dense(units= 60, activation='relu')) 
model.add(Dense(units= 72, activation='relu')) 

# 출력층 이진분류 activation : sigmoid, units :1 
# units : 퍼셉트론의 숫자(생각하는 방법의 수)> 출력해야하는 정답의 종류
# activation : 다음층에 값을 전달하는 방법 > 마지막층은 값을 종합해서 결론 
model.add(Dense(units= 1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# 학습 방법 설정
model.compile(
    loss='binary_crossentropy',
    optimizer ='adam',
    metrics=['accuracy']
)
```

```python
model.fit(X_train,y_train,epochs =100,
          #validation_data =(X_test,y_test) test데이터가 있을때
          validation_split=0.3 # test 데이터가 없을 때 
          )
```
